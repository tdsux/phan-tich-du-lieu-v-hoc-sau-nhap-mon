```python
import numpy as np
import pandas as pd

#1
data = {
    "Product": ["Laptop", "Phone", "Chair", "Desk", "Headphone", "Monitor", "Sofa", "Keyboard", "Mouse", "Tablet"],
    "Category": ["Electronics", "Electronics", "Furniture", "Furniture", "Accessories", "Electronics", "Furniture", "Accessories", "Accessories", "Electronics"],
    "Price": [1500, 800, 120, 300, 200, 600, 700, 150, 50, 400],
    "Date": ["2024-01-15", "2024-02-10", "2024-03-15", "2024-03-18", "2024-04-12", "2024-05-20", "2024-06-20", "2024-02-08", "2024-07-14", "2024-09-30"],
    "Quantity": [2, 5, 3, 1, 4, 2, 1, 6, 8, 3]
}

df = pd.DataFrame(data)
#2
df.head(5)
```

```python
#3
print("Giá trị trung bình cột Price: ", df["Price"].mean())
```

```python
#4
print("Trung bình cột Quantity nhóm theo Category:\n", df.groupby(["Category"]).mean())
```

```python
#5
print(df.sort_values(by=["Price"], ascending=False))
```

```python
#6
import matplotlib.pyplot as plt
plt.bar(df["Product"], df["Price"])
plt.title("Price chart")
plt.xlabel("Price")
plt.xticks(rotation=45)
plt.ylabel("Product")
plt.show()
```

```python
#7
plt.scatter(df["Quantity"], df["Price"])
plt.title("Biểu đồ phân tán")
plt.xlabel("Quantity")
plt.ylabel("Price")
plt.show()
```

```python
#8
plt.plot(df["Date"], df["Price"])
plt.xticks(rotation=45)
plt.show()
```

```python
#9
df9 = df.groupby(["Category"]).sum()
print(df9)

plt.bar(["Acessories", "Electronics", "Furniture"], df9["Price"])
plt.title("Price chart")
plt.xlabel("Price")
plt.ylabel("Product")
plt.show()
```

```python
#10
df9 = df.isnull().values.any()
if df9 == False:
    print("Không có giá trị NaN")
else:
    print("Có giá trị NaN")
```

```python
#11
df11 = df.drop_duplicates(subset=["Product"])
print(df11)
```

```python
#12
df12 = df.loc[(df["Quantity"] > 5)]
print(df12)
```

```python
#13
df = df.assign(Revenue=df["Quantity"] * df["Price"])
print(df)

```

```python
#14
df.loc[df['Revenue'] < 500, 'Sale_Level'] = 'Low'
df.loc[(df['Revenue'] >= 500) & (df["Revenue"] <= 2000), 'Sale_Level'] = 'Medium'
df.loc[df['Revenue'] > 2000, 'Sale_Level'] = 'High'

print(df)
```

```python
#15
import matplotlib.pyplot as plt
df15 = df.groupby(["Sale_Level"]).count()

plt.bar(["High", "Medium", "Low"], df15["Quantity"])
plt.title("Price chart")
plt.xlabel("Sale Level")
plt.xticks(rotation=45)
plt.ylabel("Product")
plt.show()
```

```python
#16
df.loc[df['Revenue'] > 1000, 'Discount'] = 0.1
df.loc[df['Revenue'] <= 1000, 'Discount'] = 0.05

print(df)
```

```python
#17
df = df.assign(Final_Revenue=df["Revenue"] - (df["Revenue"] * df["Discount"]))
print(df)
```

```python
#18
df['Month'] = pd.to_datetime(df['Date']).dt.month
print(df)
```

```python
#19

grouped_df = df.groupby('Month')['Revenue'].sum().reset_index()

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(grouped_df['Month'], grouped_df['Revenue'])
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()
```
