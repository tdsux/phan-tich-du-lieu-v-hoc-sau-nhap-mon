```python
#Phần 1: Thống kê dữ liệu
import pandas as pd
import numpy as np

duLieuXetTuyenDaiHoc = pd.read_csv("processed_dulieuxettuyendaihoc.csv")

# print(duLieuXetTuyenDaiHoc.describe())
#1
# print(duLieuXetTuyenDaiHoc.sort_values(by="DH1"))
print(duLieuXetTuyenDaiHoc["DH1"].sort_values())

#2
print(duLieuXetTuyenDaiHoc["GT"].sort_values())

#3
def q1(x):
    return x.quantile(0.25)

def q2(x):
    return x.quantile(0.50)

def q3(x):
    return x.quantile(0.75)
pivotTable1 = duLieuXetTuyenDaiHoc.pivot_table(index="KT", values="DH1", aggfunc={"DH1": ["count", "sum", "mean", "median", "min", "max", "std", q1, q2, q3]})
print(pivotTable1)

#4
pivotTable2 = duLieuXetTuyenDaiHoc.pivot_table(index=["KT", "KV"], values="DH1", aggfunc={"DH1": ["count", "sum", "mean", "median", "min", "max", "std", q1, q2, q3]})
print(pivotTable2)
```

```python
#Phan 2: Trinh bay du lieu
#1
import matplotlib.pyplot as plt

# print(duLieuXetTuyenDaiHoc["GT"].value_counts())
tanSuat = duLieuXetTuyenDaiHoc["GT"].value_counts(normalize=True)

fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.hist(duLieuXetTuyenDaiHoc["GT"])
ax1.set_title("Biểu đồ tần số")
ax1.set_xlabel("Giới tính")
ax1.set_ylabel("Tần số")


ax2.pie(tanSuat, labels=tanSuat.index, autopct='%1.1f%%', startangle=140)
ax2.set_title("Biểu đồ tần suất")

plt.show()



#2
print(duLieuXetTuyenDaiHoc[["US_TBM1", "US_TBM2", "US_TBM3"]])

#3
print(duLieuXetTuyenDaiHoc.loc[(duLieuXetTuyenDaiHoc["GT"] == "M"), "DT"])

#4
print(duLieuXetTuyenDaiHoc.loc[(duLieuXetTuyenDaiHoc["DT"] == 0) & (duLieuXetTuyenDaiHoc["DH1"] >= 5.0) & (duLieuXetTuyenDaiHoc["DH2"] >= 4.0) & (duLieuXetTuyenDaiHoc["DH3"] >= 4.0), "KV"])

#5
print(duLieuXetTuyenDaiHoc.loc[(duLieuXetTuyenDaiHoc["DH1"] >= 5.0) & (duLieuXetTuyenDaiHoc["DH2"] >= 5.0) & (duLieuXetTuyenDaiHoc["DH3"] >= 5.0) & (duLieuXetTuyenDaiHoc["KV"] == "2NT")])
```
