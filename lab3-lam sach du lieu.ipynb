```python
#Van de 1
columnNames = ["ID", "Name", "Age", "Weight", "m0006", "m0612", "m1218", "f0006", "f0612", "f1218"]
patientHeartRate = pd.read_csv("https://raw.githubusercontent.com/huynhhoc/DataAnalystDeepLearning/main/Data/patient_heart_rate.csv", names=columnNames)
patientHeartRate.head(5)
```

```python
#Vam de 2
patientHeartRate[["FirstName", "LastName"]] = patientHeartRate["Name"].str.split(expand=True)
patientHeartRate = patientHeartRate.drop("Name", axis=1)
patientHeartRate.head(5)
```

```python
#Van de 3
weight = patientHeartRate["Weight"]
for i in range (0, len(weight)):
    x = str(weight[i])
    if "lbs" in x[-3:]:
        x = x[:-3]
        floatX = float(x)
        y = int(floatX / 2.2)
        y = str(y) + "kgs"
        weight[i] = y

print(patientHeartRate)
```

```python
#Van de 4
patientHeartRate.dropna(how="all", inplace=True)
print(patientHeartRate)
```

```python
#Van de 5
patientHeartRate = patientHeartRate.drop_duplicates(subset=["FirstName", "LastName", "Age", "Weight"])
print(patientHeartRate)
```

```python
#van de 6
patientHeartRate.FirstName.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
patientHeartRate.LastName.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print(patientHeartRate)
```

```python
#van de 7
#a
patientHeartRate.dropna(how="any", inplace=True)
print(patientHeartRate)
```

```python
#van de 7
#b
patientHeartRate = patientHeartRate.fillna(0)

print(patientHeartRate)
```

```python
#van de 7
#c

patientHeartRate.fillna(patientHeartRate.mean(), inplace=True)
print(patientHeartRate)
```

```python
#van de 7
#d
patientHeartRate = patientHeartRate.fillna(patientHeartRate.mode().iloc[0])
print(patientHeartRate)
```

```python
#van de 8
patientHeartRate = pd.melt(patientHeartRate, id_vars=["ID", "Age", "Weight", "FirstName", "LastName"], value_name="PulseRate", var_name="sex_and_time").sort_values(["ID", "Age", "Weight", "FirstName", "LastName"])

tmpDf = patientHeartRate["sex_and_time"].str.extract(r"(\D)(\d+)(\d{2})", expand=True)

tmpDf.columns = ["Sex", "hours_lower", "hours_upper"]

tmpDf["Time"] = tmpDf["hours_lower"] + "-" + tmpDf["hours_upper"]

patientHeartRate = pd.concat([patientHeartRate, tmpDf], axis=1)

patientHeartRate = patientHeartRate.drop(["sex_and_time", "hours_lower", "hours_upper"], axis=1)
patientHeartRate = patientHeartRate.dropna()
patientHeartRate.to_csv("outputcleanup.csv", index=False)
print(patientHeartRate)
```
